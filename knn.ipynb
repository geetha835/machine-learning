```python
import pandas as pd
df=pd.read_csv("/home/rgukt-basar/MachineLearning/train_titanic.csv")
df
```

```python
df['Age']=df['Age'].fillna(df['Age'].median())
df
```

```python
most_frequent_embarked = df["Embarked"].mode()[0]
print(most_frequent_embarked)
```

```python
df['Embarked']=df['Embarked'].fillna(df['Embarked']=='S')
df
```

```python
df=df.drop(columns=['Name','Ticket','Cabin'])
df
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
df['Sex']=label.fit_transform(df['Sex'])
df=pd.get_dummies(df,columns=['Embarked'],drop_first=True).replace({True:1,False:0})
df
```

```python
df=df.drop(columns=['Embarked_C'])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
x=df[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare','Embarked_Q', 'Embarked_S']]
y=df['Survived']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python

from sklearn.impute import SimpleImputer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Try different values of k
k_values = range(1, 21)
accuracies = []
#accuracies = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    y_pred = knn.predict(x_test)
    accuracies.append(accuracy_score(y_test, y_pred))

# Find the best k
best_k = k_values[np.argmax(accuracies)]
best_accuracy = max(accuracies)

# Plot accuracy vs k-values
plt.figure(figsize=(8, 5))
plt.plot(k_values, accuracies, marker='o', linestyle='dashed', color='b')
plt.xlabel("Number of Neighbors (k)")
plt.ylabel("Accuracy")
plt.title("KNN Accuracy for Different k Values")
plt.xticks(k_values)
plt.grid()
plt.show()

print(f"Best k value: {best_k}, Best Accuracy: {best_accuracy:.2%}")
```

```python
# Train final KNN model with best k
knn_best = KNeighborsClassifier(n_neighbors=best_k)
knn_best.fit(x_train, y_train)
y_pred_best = knn_best.predict(x_test)

# Evaluate model
final_accuracy = accuracy_score(y_test, y_pred_best)
final_report = classification_report(y_test, y_pred_best)

print(f"Final Accuracy: {final_accuracy:.2%}")
print("Classification Report:\n", final_report)
```
